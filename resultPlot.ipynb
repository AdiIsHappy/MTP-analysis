```python
import os
import glob
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
ratios_data = pd.read_csv(os.path.abspath('./results/ratios.csv'))
ratios_data_dur = pd.read_csv(os.path.abspath('./results/ratios_dur.csv'))
groups=pd.read_csv(os.path.abspath('./Group.csv'),header=0)
```

```python
ratio_copy=pd.merge(ratios_data, groups, on='id', how='left')
ratio_copy=pd.merge(ratio_copy, ratios_data_dur, on='id', how='left')
```

```python
numeric_columns = ratio_copy.select_dtypes(include=[np.number])
ratios_grouped = numeric_columns.groupby(ratio_copy['Group']).mean()
ratios_grouped.columns
```

### Alpha/Beta Ratio (ABR) Relaxation

```python
# plt.bar(["Alpha/Beta Ratio dur", "Alpha/Beta Ratio pre", "Alpha/Beta Ratio post"], ratios_grouped[["Alpha/Beta Ratio dur", "Alpha/Beta Ratio pre", "Alpha/Beta Ratio post"]])
# Plot Alpha/Beta Ratio for dur, pre and post in a side-by-side bar graph
x = np.arange(len(ratios_grouped.index))  # the label locations
width = 0.12  # the width of the bars

plt.figure(figsize=(12, 6))
plt.bar(x - 2*width, ratios_grouped["Alpha/Beta Ratio pre"], width, label="pre Simulation", alpha=0.7)
plt.bar(x - width, ratios_grouped["Alpha/Beta Ratio durPre"], width, label="simulation pre earthquake", alpha=0.7)
plt.bar(x , ratios_grouped["Alpha/Beta Ratio durEarthquake"], width, label="Simulation during earthquake", alpha=0.7)
plt.bar(x + width, ratios_grouped["Alpha/Beta Ratio durPost"], width, label="Simulation post earthquake", alpha=0.7)
plt.bar(x + 2*width, ratios_grouped["Alpha/Beta Ratio post"], width, label="post Simulation", alpha=0.7)


plt.xlabel("Group")
plt.ylabel("Alpha/Beta Ratio")
plt.title("Alpha/Beta Ratio: Higher ABR = relaxation")
plt.xticks(x, ratios_grouped.index)

plt.legend()
plt.tight_layout()
plt.show()
```
